```
import random

def play_21():
    print("Добро пожаловать в игру '21 очко'!")

    #Получение карт
    player_cards=[random.randint(2, 11), random.randint(2, 11)]
    dealer_cards=[random.randint(2, 11)]

    player_score=sum(player_cards)
    dealer_score=sum(dealer_cards)

    print(f'\nВаши карты: {player_cards}, сумма: {player_score}')
    print(f'Карта дилера: {dealer_cards}, сумма: {dealer_score}')

    #Ход игрока
    while player_score<=21:
        choice=input('\nХотите взять карту? (да/нет): ').lower()
        if choice=='да':
            new_card = random.randint(2, 11)
            player_cards.append(new_card)
            player_score +=new_card
            print(f'\nВы получили: {new_card}')
            print(f'Ваши карты: {player_cards}, сумма: {player_score}')
        elif choice=='нет':
            break
        else:
            print("Введите 'да' или 'нет'")

    if player_score>21:
        print('\nПеребор! Вы проиграли!')
        return

    #Ход дилера
    dealer_cards.append(random.randint(2, 11))
    dealer_score=sum(dealer_cards)

    print(f"\nКарты дилера: {dealer_cards}, сумма: {dealer_score}")

    while dealer_score<17:
        new_card=random.randint(2, 11)
        dealer_cards.append(new_card)
        dealer_score +=new_card
        print(f"Дилер взял: {new_card}")
        print(f"Карты дилера: {dealer_cards}, сумма: {dealer_score}")

    #Определение победителя
    print("\nРезультат:")
    if dealer_score > 21:
        print('Дилер перебрал карт! Вы выиграли!')
    elif player_score > dealer_score:
        print('Вы выиграли!')
    elif player_score== dealer_score:
        print('Ничья')
    else:
        print('Вы проиграли')

play_21()
```

```
import random
import time
import json
from datetime import datetime

def guess_number():
    print("Добро пожаловать в игру 'Угадай число'!")
    print("Я загадал число от 1 до 100. Попробуй угадать!")

    # Загадываем число
    secret_number=random.randint(1, 100)
    attempts=0
    start_time=time.time()

    while True:
        try:
            guess=int(input("Введите догадку (1-100): "))
            attempts+=1

            if guess<secret_number:
                print('Загаданное число больше')
            elif guess>secret_number:
                print('Загаданное число меньше')
            else:
                end_time=time.time()
                game_time=round(end_time - start_time, 2)
                print(f'Поздравляю! Вы угадали число за {attempts} попыток и {game_time} секунд!')
                return True, attempts, game_time

        except ValueError:
            print('Введите целое число от 1 до 100')

def save_statistics(result, attempts, game_time):
    try:
        #Загружаем существующую статистику
        with open('guess_number_stats.json', 'r') as f: #открыть в режиме чтения
            stats = json.load(f)
    except (FileNotFoundError, json.JSONDecodeError):
        #Если нет - создаю новую статистику
        stats = []

    #Новая запись
    stats.append({'date': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),'result': 'Победа' if result else 'Поражение',
        'attempts': attempts,
        'time_seconds': game_time})

    #Сохраняем обновленную статистику
    with open('guess_number_stats.json', 'w') as f:
        json.dump(stats, f, indent=4, ensure_ascii=False)

def main():
    while True:
        result, attempts, game_time = guess_number()
        save_statistics(result, attempts, game_time)

        play_again = input("Хотите сыграть еще раз? (да/нет): ").lower()
        if play_again not in ['да', 'д', 'yes', 'y']:
            print("Спасибо за игру! Статистика сохранена")
            break

if __name__ == "__main__":
    main()
```

```
def display_board(board):
    for i in range(3):
        print(f' {board[i*3]} | {board[i*3+1]} | {board[i*3+2]} ')
        if i<2:
            print("-----------")

def display_instructions():
    print('Выберите клетку для хода (номера от 1 до 9):')
    print('1 | 2 | 3')
    print('4 | 5 | 6')
    print('7 | 8 | 9')
    print()

def check_winner(board):
    #Проверка, есть ли победитель или ничья
    win_lines = [
        [0,1,2], [3,4,5], [6,7,8],  #строки
        [0,3,6], [1,4,7], [2,5,8],  #столбцы
        [0,4,8], [2,4,6]            #диагонали
    ]

    for line in win_lines:
        if board[line[0]] == board[line[1]] == board[line[2]] != ' ':
            return board[line[0]]

    return 'Ничья' if ' ' not in board else None

def get_player_move(board, current_player):
    #Получает и проверяет ход игрока
    while True:
        move=input(f'Ход игрока {current_player} (1-9):')
        try:
            move=int(move) - 1  #преобразуем в индекс (0-8)
            if not 0 <= move <= 8:
                print('Введите число от 1 до 9!')
                continue

            if board[move]!=' ':
                print('Эта клетка уже занята!')
                continue
            return move

        except ValueError:
            print('Введите число от 1 до 9!')

#Основная функция игры
def play_game():

    board=[' '] * 9
    current_player='X'

    display_instructions() #подпись клеток

    while True:
        display_board(board)
        move=get_player_move(board, current_player)
        board[move]=current_player #запись хода

        winner=check_winner(board) #проверка 3х одинак символов
        if winner:
            display_board(board)
            if winner=='Ничья':
                print('Ничья!')
            else:
                print(f'Игрок {winner} победил!')
            break
        #смена игрока
        current_player = 'O' if current_player == 'X' else 'X'

def ask_to_play_again():
    while True:
        answer = input('\nХотите сыграть ещё раз? (да/нет): ').lower()
        if answer in ('да', 'д', 'yes', 'y'):
            return True
        elif answer in ('нет', 'н', 'no', 'n'):
            return False
        print('Пожалуйста, введите "да" или "нет"')

def main():
    print('Добро пожаловать в Крестики-Нолики!')
    while True:
        play_game()
        if not ask_to_play_again():
            print('Спасибо за игру! До свидания!')
            break

if __name__ == '__main__':
    main()
```
